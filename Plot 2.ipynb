```python
# Import necessary packages
import pandas as pd
import datetime as dt
import matplotlib.pylab as plt
%matplotlib inline
```

```python
# Read the .csv file
boarddf = pd.read_csv('acoustic_features.csv')
boarddf.head()
```

```python
# Change date column type to datetime
boarddf['date'] = pd.to_datetime(boarddf['date'], format='%Y-%m-%d')

# Set  datetime index
boarddf.set_index('date', inplace = True)
```

```python
# Sort boarddf by datetime index
boarddf = boarddf.sort_index()
```

```python
# Select data from 10's
boarddf_10_19 = boarddf['2010':'2019']
```

```python
# Plot
acous = boarddf_10_19.groupby('artist')['acousticness'].mean()
dance = boarddf_10_19.groupby('artist')['danceability'].mean()
x = acous
y = dance

plt.scatter(x, y)
plt.title('Scatter Plot of Danceability and Acousticness')
plt.xlabel('Acousticness')
plt.ylabel('Danceability')
plt.show()
```
